# RATESYLLABLES

```python
import pandas as pd
import numpy as np
import json
import glob
import seaborn as sb
from matplotlib import pyplot
```

```python
def getsylldata(f):
    f['responses']=f['responses'].astype(str) # make everything in responses as string
    f.loc[f['responses']=='nan','responses']='x' # put a dummy value for NaNs
    resploc = pd.Series(list(f['responses'])).str.contains('Q0') # locations of all responses
    allresps = list(f.loc[resploc,'responses'])
    assert len(allresps)==23, 'Incorrect num of responses' # Make sure 23 responses
    # First extract person code and languages
    curr_sscode = f['subject'][0]
    curr_sslang = json.loads(allresps[0])['Q0']
    
    # Set up  columns for output
    colS=[]
    colL=[]
    col1=[]
    col2=[]
    col3=[]
    colIP=[] # added column for IP
    colRT=[] # added column for RT
        
    allresps=allresps[1:] # First one is ss_lang
    dum=pd.Series(list(f.columns.astype(str))) # Mostly to check if IP field exists
    
    for curr_resp in allresps:
        curr_idx = f.index[pd.Series(list(f['responses'])).str.contains(curr_resp)][0]
        curr_stim = f.loc[curr_idx+1,'stimulus'][4:]
        curr_qlty = chr(f.loc[curr_idx+1,'key_press'].astype(int))
        curr_RT = f.loc[curr_idx+1,'rt']
        if any(dum.isin(['IP'])):
            curr_IP = f.loc[curr_idx,'IP']
        else:
            curr_IP = '0.0.0.0'
        # Save
        colS.append(curr_sscode)
        colL.append(curr_sslang)
        col1.append(curr_stim)
        col2.append(json.loads(curr_resp)['Q0'])
        col3.append(curr_qlty)
        colIP.append(curr_IP)
        colRT.append(curr_RT)
    
    outdf = pd.DataFrame({'SsCode':colS, 'IP':colIP, 'Languages':colL, 'Stimulus':col1, 'Response':col2, 'Rating':col3, 'RT':colRT})
    return outdf
```

## Main loop to collect data

```python
# Actual code
allfiles = glob.glob('MT*.csv')
rate_syll_results=pd.DataFrame([]) # start with empty
allIP=[] # Store IPs for quality control
response_df = pd.DataFrame([])
for fname in allfiles:
    f = pd.read_csv(fname)
    dum=pd.Series(list(f.columns.astype(str)))
    #print(fname) #*** Uncomment to debug
    f_df = getsylldata(f)

    # Get responses for easy manual checking
    dum2 = pd.DataFrame({'Stim':f_df['Stimulus'], fname:f_df['Response']})
    dum2.sort_values(by=['Stim'],inplace=True)
    dum2 = dum2.reset_index(drop=True)
    response_df = pd.concat([response_df,dum2[fname]],axis=1,sort=False, ignore_index=False)

    # IP:
    if any(dum.isin(['IP'])):
        allIP.append(f.loc[1,'IP'])
    else:
        allIP.append('0.0.0.0')
        
    if rate_syll_results.shape[1]>2:
        f_df = f_df[list(f_df.columns)[0:]] # keep only last two - assume everything is ok!
        rate_syll_results = pd.concat([rate_syll_results, f_df], axis=0, sort=False)
    else:
        rate_syll_results = pd.concat([rate_syll_results, f_df])
print('Done!')
```

## Make csv for easy detection of bad data

```python
response_df = pd.concat([pd.DataFrame({'Filename':dum2['Stim']}),response_df],axis=1, sort=False)
response_df.T.to_csv('Responses4ManualChecking.csv', header=False)
```

## Get some aggregate numbers in readable files

```python
allstims = rate_syll_results['Stimulus'].unique()
aggr_data = pd.DataFrame([]);
for currstim in allstims:
    dum=[]
    dum.append(currstim)
    dum.append(rate_syll_results.loc[rate_syll_results['Stimulus']==currstim,'Rating'].mean())
    dum.append(rate_syll_results.loc[rate_syll_results['Stimulus']==currstim,'Rating'].median())
    dum.append(rate_syll_results.loc[rate_syll_results['Stimulus']==currstim,'Response'].value_counts().idxmax())
    dum.append(rate_syll_results.loc[rate_syll_results['Stimulus']==currstim,'Response'].value_counts()[0])
    dum.append(rate_syll_results.loc[rate_syll_results['Stimulus']==currstim,'Response'].value_counts())
    aggr_data = pd.concat([aggr_data, pd.DataFrame(dum)],axis=1)
aggr_data.to_csv('aggrdata_all.csv')
```
